```python
import matplotlib.pyplot as plt
import seaborn as sns

I1=[1.5,6.1,12.6,19.5,28.9]
U1=[5.79,5.08,4.04,2.94,1.35]

I2=[2.4,7.4,21.2,25.6,31.7]
U2=[5.69,4.93,2.84,2.17,1.24]

I3=[6.0,11.1,17.5,22.3,31.2]
U3=[5.03,4.24,3.28,2.55,1.20]

sns.set_style("whitegrid")

sns.scatterplot(x=I1, y=U1)
sns.lineplot(x=I1, y=U1,label='U-I')

sns.scatterplot(x=I2, y=U2, color='red')
sns.lineplot(x=I2, y=U2, color='red',label='U\'-I\'')

sns.scatterplot(x=I3, y=U3, color='green')
sns.lineplot(x=I3, y=U3, color='green',label='U\'\'-I\'\'')


plt.xlim(0,33)
plt.ylim(0,6)
plt.legend()
# Set the color of the axes to black
ax = plt.gca()
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')

# Add arrows at the ends of the axis
plt.annotate('', xy=(33, 0), xytext=(0, 0), arrowprops=dict(arrowstyle='->', color='black'))
plt.annotate('', xy=(0, 6), xytext=(0, 2), arrowprops=dict(arrowstyle='->', color='black'))

# Add x labels
plt.text(33, -0.5, 'I(mA )', ha='right')
plt.text(33, 0.2, 'I\'(mA )', ha='right')
plt.text(33, 0.65, 'I\'\'(mA )', ha='right')

# Add y labels
plt.text(0.1, 5.9, 'U(V )', va='top')
plt.text(0.1, 5.5, 'U\'(V )', va='top')
plt.text(0.1, 5.1, 'U\'\'(V )', va='top')

plt.show()
```

```python
import numpy as np

# Fit the data to a line
coeffs1 = np.polyfit(I1, U1, 1)
coeffs2 = np.polyfit(I2, U2, 1)
coeffs3 = np.polyfit(I3, U3, 1)

# The slope and intercept of the fitted line are
slope1, intercept1 = coeffs1
slope2, intercept2 = coeffs2
slope3, intercept3 = coeffs3

print("For the first set of data, the slope is {:.4f} and the intercept is {:.4f}".format(slope1, intercept1))
print("For the second set of data, the slope is {:.4f} and the intercept is {:.4f}".format(slope2, intercept2))
print("For the third set of data, the slope is {:.4f} and the intercept is {:.4f}".format(slope3, intercept3))
```
