```python
import matplotlib.pyplot as plt
import numpy as np

pi=3.1415926535
U=[226.3,226.0,225.8]

I_C=[70.1*10**-3,212.9*10**-3,492.1*10**-3]
I_RL=[295.1*10**-3,296.01*10**-3,296.0*10**-3]
phi=[58,26,298]

phi[2]=360-phi[2]

U=[U/1000 for U in U]
I_C=[I_C for I_C in I_C]
I_RL=[I_RL for I_RL in I_RL]
phi_rad = [np.radians(p) for p in phi]

alpha=[]
for phi,ic,irl in zip(phi_rad,I_C,I_RL):
    temp_theta=np.arcsin(ic*np.sin(pi/2+phi)/irl)
    alpha.append(pi-temp_theta-phi-pi/2)

# 定义相量
vectors_U= [
    {"name": "v1", "vector": np.array([U[0], 0])},
    {"name": "v2", "vector": np.array([U[1], 0])},
    {"name": "v3", "vector": np.array([U[2], 0])},
]

vectors_I_C= [
    {"name": "v1", "vector": np.array([0, I_C[0]])},
    {"name": "v2", "vector": np.array([0, I_C[1]])},
    {"name": "v3", "vector": np.array([0, I_C[2]])},
]

vectors_I_RL= [
    {"name": "v1", "vector": np.array([I_RL[0]*np.sin(alpha[0]), -I_RL[0]*np.cos(alpha[0])])},
    {"name": "v2", "vector": np.array([I_RL[1]*np.sin(alpha[1]), -I_RL[1]*np.cos(alpha[1])])},
    {"name": "v3", "vector": np.array([-I_RL[2]*np.sin(alpha[2]), -I_RL[2]*np.cos(alpha[2])])},
]

# 创建一个新的图形
plt.figure()

# 画出每一个相量
f=1
for v1,v2,v3 in zip(vectors_U,vectors_I_C,vectors_I_RL):
    if f==1:
        plt.quiver(0, 0, v1["vector"][0], v1["vector"][1], angles='xy', scale_units='xy', scale=1, label=r"$C=1\mu F$")
        plt.annotate(r'$\dot{U}$,226.0V', (v1["vector"][0], v1["vector"][1]))
        plt.quiver(0, 0, v2["vector"][0], v2["vector"][1], angles='xy', scale_units='xy', scale=1)
        plt.annotate(r'$I_{C1}.0.235A$', (v2["vector"][0]-0.05, v2["vector"][1]))
        plt.quiver(v2["vector"][0], v2["vector"][1], v3["vector"][0], v3["vector"][1], angles='xy', scale_units='xy', scale=1)
        plt.quiver(0, 0, v3["vector"][0], v3["vector"][1]+v2["vector"][1], angles='xy', scale_units='xy', scale=1)
        plt.annotate(r'$I_1$', (v3["vector"][0]-0.02, v3["vector"][1]+v2["vector"][1]-0.03))
        
    if f==2:
        plt.quiver(0, 0, v1["vector"][0], v1["vector"][1], angles='xy', scale_units='xy', scale=1, color='r',label=r"$C=3\mu F$")
        plt.quiver(0, 0, v2["vector"][0], v2["vector"][1], angles='xy', scale_units='xy', scale=1, color='r')
        plt.annotate(r'$I_{C2},0.144A$', (v2["vector"][0]-0.05, v2["vector"][1]))
        plt.quiver(v2["vector"][0], v2["vector"][1], v3["vector"][0], v3["vector"][1], angles='xy', scale_units='xy', scale=1, color='r')
        plt.quiver(0, 0, v3["vector"][0], v3["vector"][1]+v2["vector"][1], angles='xy', scale_units='xy', scale=1, color='r')
        plt.annotate(r'$I_2$', (v3["vector"][0]-0.02, v3["vector"][1]+v2["vector"][1]-0.03))
        
    if f==3:
        plt.quiver(0, 0, v1["vector"][0], v1["vector"][1], angles='xy', scale_units='xy', scale=1, color='b',label=r"$C=6.7\mu F$")
        plt.quiver(0, 0, v2["vector"][0], v2["vector"][1], angles='xy', scale_units='xy', scale=1, color='b')
        plt.annotate(r'$I_{C3},0.492A$', (v2["vector"][0]-0.05, v2["vector"][1]))
        plt.quiver(v2["vector"][0], v2["vector"][1], v3["vector"][0], v3["vector"][1], angles='xy', scale_units='xy', scale=1, color='b')
        plt.quiver(0, 0, v3["vector"][0], v3["vector"][1]+v2["vector"][1], angles='xy', scale_units='xy', scale=1, color='b')    
        plt.annotate(r'$I_3$', (v3["vector"][0]-0.02, v3["vector"][1]+v2["vector"][1]-0.08))
    f+=1

# 设置图形的比例
plt.xlim(-0.1,0.4)
plt.ylim(-0.3,0.55)
# 隐藏坐标轴
plt.axis('off')
# 添加图例
plt.legend()

# 显示图形
plt.show()
```
